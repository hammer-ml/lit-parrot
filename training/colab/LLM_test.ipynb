```python
import json 
file_path="./databricks-dolly-15k.jsonl"
with open(file_path, "r") as file:
    data = file.readlines()
    data = [json.loads(line) for line in data]
for item in data:
    item["input"] = item.pop("context")
    item["output"] = item.pop("response")
```
